```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
d=pd.read_csv("diabetes.csv")
d
```

```python
feature_cols=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
x=d[feature_cols]
y=d.Outcome
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
lin=LogisticRegression()
lin.fit(xtrain,ytrain)
```

```python
y_predict=lin.predict(xtest)
y_predict
```

```python
cm=metrics.confusion_matrix(ytest,y_predict)
cm
```

```python
print("Accuracy:",metrics.accuracy_score(ytest,y_predict))
print("Precision:",metrics.precision_score(ytest,y_predict))
print("recall:",metrics.recall_score(ytest,y_predict))
```

```python
class_names=[0,1]
fig,ax=plt.subplots()
tick_marks=np.arange(len(class_names))
plt.xticks(tick_marks,class_names)
sb.heatmap(pd.DataFrame(cm),annot = True,cmap = 'Purples',fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.xlabel("predicted values")
plt.ylabel("Actual values")
plt.title("confusion matrix")
plt.show()
```

```python
y_predict_proba = lin.predict_proba(xtest)[::,1]
fpr,tpr,_= metrics.roc_curve(ytest,y_predict_proba)
auc = metrics.roc_auc_score(ytest,y_predict_proba)
plt.plot(fpr,tpr,label = "data 1,auc ="+str(auc))
plt.legend(loc = 'best')
plt.show()
```
