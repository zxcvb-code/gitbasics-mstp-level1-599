```python
t=sb.load_dataset("tips")
t
```

```python
sb.factorplot(x="total_bill",y="day",data=t)
```

```python
sb.factorplot(x="total_bill",y="day",hue="sex",markers=['x','s'],data=t)
```

```python
sb.factorplot(x="total_bill",y="day",hue="sex",markers=['x','s'],data=t,dodge=True)
```

```python
from numpy import median
sb.factorplot(x="total_bill",y="day",hue="sex",markers=['x','s'],data=t,estimator=median)
```

```python
sb.factorplot(x="sex",y="tip",hue="day",markers=['d','x','d','s'],data=t,estimator=median)
```

```python
help(sb.pointplot)
```

```python
sb.pointplot(x="total_bill",y="size",data=t)
```

```python
sb.pointplot(x="total_bill",y="size",data=t,palette="copper")
```

```python
sb.pointplot(x="total_bill",y="size",data=t,markers=["x",'s'],linestyles=["-.","--"])
```

```python
sb.pointplot(x="tip",y="day",data=t,hue="sex",color="g")
```

```python
sb.pointplot(x="tip",y="day",data=t,hue="sex",join=False,dodge=True)
```

```python
sb.pointplot(x="tip",y="day",data=t,hue="sex",join=False,dodge=True,estimator=median)
```
