```python
## scatter plot
h=list(range(1,10,1))
s=list(range(2,20,2))
w.scatter(h,s,color='pink')#r,g,w,b,pink
w.xlabel('x-axis')
w.ylabel('y-axis')
w.title('scatter plot')
w.show()
```

```python
# draw a plot for time and distance by using list comprehension 
#distance=2*time
h=[t for t in range(1,10,1)]
b=[t*2 for t in range(2,20,2)]
w.xlabel('time')
w.ylabel('distance')
w.title('line plot')
w.plot(h,b)
w.show()
```

```python
import matplotlib.pyplot as w
x=list(range(5,51,5))
y=list(range(1,11,1))
w.plot(y,x)
w.title('A graph is a example to ticks')
w.xlabel('natural number')
w.ylabel('values with ++5 up to 50')
w.show()
```

```python
import matplotlib.pyplot as w
import numpy as n
x=[5,10,15,20,25,30,35,40,45,50]
print(x)
y=[3,4,5,2,6,1,8,7,9,2]
print(y)
w.plot(x,y)
w.title('graph is example for ticks')
w.xlabel('natural number')
w.ylabel('values with ++5 up to 50')
w.show()
```

```python
import matplotlib.pyplot as w
import numpy as n
w.xticks(n.arange(1,11,1))
w.yticks(n.arange(5,51,5))
w.plot(x,y)
print(x)
print(y)
w.title("generate the values by using the using and numpy")
w.xlabel("natural number")
w.ylabel("values seperate with 5 up to 50")
w.show()
```

```python
import numpy as n
import matplotlib.pyplot as w
time=n.arange(1,10,1)
amptitude=n.sin(time)
w.plot(time,amptitude)
w.title('sine wave A=sin(wt)')
w.xlabel('time')
w.ylabel('amptitude')
w.show()
```

```python
file="income.csv"
import pandas as p
d=p.read_csv(file)
d
```
