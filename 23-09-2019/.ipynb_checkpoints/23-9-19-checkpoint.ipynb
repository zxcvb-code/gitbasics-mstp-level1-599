# conditional statements 
#### 1.if----if(condition)------>true only
#### 2.if-else-----if(condition) statements  else(condition) statements------->true/false
#### 3.nested if/if-else/else-if


## 1.if-else

```python
c=8
if c>0:                                # : must kept after condition
    print("c is greater")              # "whitespace"should be kept before stmts
else:
    print("c is lesser")
```

```python
s="python"
c=input()
if(s==c):
    print("two strings are equal")
else:
    print("two strings are not equal")
```

## 2.else-if(elif)

```python
k=45
if(k!=45):
    print("not equal")
elif(k==45):
    print("equal")
else:
    print("error")
```

```python
#task
#write a program to find the number is even/odd/composite
s=int(input())
if(s==0):
    print("composite number")
elif(s%2==0):
    print("even number")
else:
    print("odd number")
```

```python
#checking the elements using if
g="jupyter"
if("y" in g):
    print("yes")
else:
    print("no")
```

## while

```python
s=0
i=0
while(i<=9):
    s=s+i
    i=i+1
    print("lop",s)
print("sum",s)
```

```python
d=int(input())
a=9
b=8
while(a>b):
    if(d==1):
        print("addition:",a+b)
    elif(d==2):
        print("sub:",a-b)
    elif(d==3):
        print("mul:",a*b)
```
