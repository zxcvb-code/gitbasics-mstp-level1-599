```python
import pandas as pd
f1=pd.DataFrame(np.random.rand(5,6),columns=['maths','science','telugu','english','physics','social'],
                index=['sukanya','gayathri','bindu','bhavya','lakshmi'])
print(f1)
print(" Total sum\n",f1.sum(axis=1))
print("AVERAGE\n",f1.mean(axis=1))
print("maximum\n",f1.max(axis=1))
print("minimum\n",f1.min(axis=1))
print("null values",f1.isna())
print(plt.plot(f1))
```
